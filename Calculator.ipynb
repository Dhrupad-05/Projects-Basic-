```python
import math

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    return x / y

def power(x, y):
    return x ** y

def square_root(x):
    if x < 0:
        return "Error! Cannot calculate square root of negative number."
    return math.sqrt(x)

def logarithm(x, base=10):
    if x <= 0:
        return "Error! Logarithm undefined for non-positive numbers."
    if base == 10:
        return math.log10(x)
    elif base == math.e:
        return math.log(x)
    else:
        return math.log(x, base)

def sine(x, degrees=True):
    if degrees:
        x = math.radians(x)
    return math.sin(x)

def cosine(x, degrees=True):
    if degrees:
        x = math.radians(x)
    return math.cos(x)

def tangent(x, degrees=True):
    if degrees:
        x = math.radians(x)
    return math.tan(x)

def cosecant(x, degrees=True):
    if degrees:
        x = math.radians(x)
    sin_val = math.sin(x)
    if sin_val == 0:
        return "Error! Cosecant undefined (sin = 0)."
    return 1 / sin_val

def secant(x, degrees=True):
    if degrees:
        x = math.radians(x)
    cos_val = math.cos(x)
    if cos_val == 0:
        return "Error! Secant undefined (cos = 0)."
    return 1 / cos_val

def cotangent(x, degrees=True):
    if degrees:
        x = math.radians(x)
    tan_val = math.tan(x)
    if tan_val == 0:
        return "Error! Cotangent undefined (tan = 0)."
    return 1 / tan_val

def factorial(n):
    if n < 0:
        return "Error! Factorial not defined for negative numbers."
    if not isinstance(n, int) and not n.is_integer():
        return "Error! Factorial only defined for integers."
    return math.factorial(int(n))

def get_pi():
    return math.pi

def calculator():
    print("=" * 50)
    print("        SCIENTIFIC CALCULATOR")
    print("=" * 50)
    print("\nOperations:")
    print("1.  Addition (+)")
    print("2.  Subtraction (-)")
    print("3.  Multiplication (×)")
    print("4.  Division (÷)")
    print("5.  Power (x^y)")
    print("6.  Square Root (√)")
    print("7.  Logarithm (log)")
    print("8.  Sine (sin)")
    print("9.  Cosine (cos)")
    print("10. Tangent (tan)")
    print("11. Cosecant (cosec)")
    print("12. Secant (sec)")
    print("13. Cotangent (cot)")
    print("14. Factorial (!)")
    print("15. Pi (π)")
    print("16. Exit")
    print("=" * 50)

    while True:
        choice = input("\nEnter your choice (1-16): ")

        if choice == '16':
            print("\nThank you for using the Scientific Calculator!")
            break

        try:
            if choice == '1':
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
                print(f"Result: {num1} + {num2} = {add(num1, num2)}")

            elif choice == '2':
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
                print(f"Result: {num1} - {num2} = {subtract(num1, num2)}")

            elif choice == '3':
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
                print(f"Result: {num1} × {num2} = {multiply(num1, num2)}")

            elif choice == '4':
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
                result = divide(num1, num2)
                print(f"Result: {num1} ÷ {num2} = {result}")

            elif choice == '5':
                base = float(input("Enter base: "))
                exponent = float(input("Enter exponent: "))
                print(f"Result: {base}^{exponent} = {power(base, exponent)}")

            elif choice == '6':
                num = float(input("Enter number: "))
                result = square_root(num)
                print(f"Result: √{num} = {result}")

            elif choice == '7':
                num = float(input("Enter number: "))
                base_choice = input("Enter base (10 for log10, e for ln, or custom): ")
                if base_choice.lower() == 'e':
                    result = logarithm(num, math.e)
                    print(f"Result: ln({num}) = {result}")
                elif base_choice == '10':
                    result = logarithm(num, 10)
                    print(f"Result: log10({num}) = {result}")
                else:
                    base = float(base_choice)
                    result = logarithm(num, base)
                    print(f"Result: log{base}({num}) = {result}")

            elif choice == '8':
                angle = float(input("Enter angle: "))
                unit = input("Degrees or Radians? (d/r): ").lower()
                result = sine(angle, unit == 'd')
                print(f"Result: sin({angle}) = {result}")

            elif choice == '9':
                angle = float(input("Enter angle: "))
                unit = input("Degrees or Radians? (d/r): ").lower()
                result = cosine(angle, unit == 'd')
                print(f"Result: cos({angle}) = {result}")

            elif choice == '10':
                angle = float(input("Enter angle: "))
                unit = input("Degrees or Radians? (d/r): ").lower()
                result = tangent(angle, unit == 'd')
                print(f"Result: tan({angle}) = {result}")

            elif choice == '11':
                angle = float(input("Enter angle: "))
                unit = input("Degrees or Radians? (d/r): ").lower()
                result = cosecant(angle, unit == 'd')
                print(f"Result: cosec({angle}) = {result}")

            elif choice == '12':
                angle = float(input("Enter angle: "))
                unit = input("Degrees or Radians? (d/r): ").lower()
                result = secant(angle, unit == 'd')
                print(f"Result: sec({angle}) = {result}")

            elif choice == '13':
                angle = float(input("Enter angle: "))
                unit = input("Degrees or Radians? (d/r): ").lower()
                result = cotangent(angle, unit == 'd')
                print(f"Result: cot({angle}) = {result}")

            elif choice == '14':
                num = float(input("Enter a non-negative integer: "))
                result = factorial(num)
                print(f"Result: {int(num)}! = {result}")

            elif choice == '15':
                print(f"Result: π = {get_pi()}")

            else:
                print("Invalid choice! Please select a number between 1 and 16.")

        except ValueError:
            print("Invalid input! Please enter valid numbers.")
        except Exception as e:
            print(f"An error occurred: {e}")

if __name__ == "__main__":
    calculator()
```
